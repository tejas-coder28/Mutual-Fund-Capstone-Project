# 02 — Data Cleaning

This notebook covers cleaning, merging, and preparing the raw CSVs for analysis.

```python
import os
from pathlib import Path

def _find_project_root(marker="data/db"):
    p = Path(os.getcwd())
    for _ in range(6):
        if (p / marker).exists():
            return p
        p = p.parent
    raise RuntimeError("Cannot locate project root. Run notebook from mf_project/")

_ROOT   = _find_project_root()
db_path = str(_ROOT / "data" / "db" / "bluestock_mf.db")
if not Path(db_path).exists() or Path(db_path).stat().st_size == 0:
    raise FileNotFoundError(
        f"DB not found or empty: {db_path}\n"
        "Run: python scripts/rebuild_db.py"
    )

import subprocess, sys

raw_csvs = list((_ROOT / 'data' / 'raw').glob('*.csv'))

if raw_csvs:
    print(f'Found {len(raw_csvs)} raw CSVs — running full cleaning pipeline...')
    subprocess.run(
        [sys.executable, str(_ROOT / 'scripts' / 'data_cleaning_and_loading.py')],
        check=True
    )
else:
    print('No raw CSVs in data/raw/ — rebuilding DB from existing processed CSVs...')
    subprocess.run(
        [sys.executable, str(_ROOT / 'scripts' / 'rebuild_db.py')],
        check=True
    )
    print('DB ready. Processed CSVs already exist in data/processed/')
```
